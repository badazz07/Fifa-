```python
import pandas as pd
fifa=pd.read_csv('fifa21_raw_data.csv')
fifa
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def remove_special_chars(text):
    return text.replace('★', '')
```

```python
fifa['IR'] = fifa['IR'].apply(remove_special_chars)
```

```python
fifa.rename(columns = {'IR':'IR(★)'}, inplace = True)
```

```python
fifa.rename(columns = {'W/F':'W/F(★)'}, inplace = True)
```

```python
fifa['W/F(★)'] = fifa['W/F(★)'].apply(remove_special_chars)
```

```python
fifa.rename(columns = {'SM':'SM(★)'}, inplace = True)
```

```python
fifa['SM(★)'] = fifa['SM(★)'].apply(remove_special_chars)
```

```python
def remove_text(text):
    return text.replace('lbs', '')
```

```python
fifa['Weight'] = fifa['Weight'].apply(remove_text)
```

```python
def clean_column_values(x):
    x = str(x)  # Ensure x is a string
    x = x.replace('€', '').replace('.', ',')  # Remove € and .
    x = x.replace('M', '00,000').replace('K', ',000')  # Replace M and K
    return x
```

```python
fifa['Value'] = fifa['Value'].apply(clean_column_values)
```

```python
fifa.rename(columns = {'Value':'Value(€)'}, inplace = True)
```

```python
def clean_wage(x):
    x = str(x)  # Ensure x is a string
    x = x.replace('€', '') # Remove € 
    x = x.replace('K', ',000')  # Replace K
    return x
```

```python
fifa['Wage'] = fifa['Wage'].apply(clean_wage)
```

```python
fifa.rename(columns = {'Wage':'Wage(€)'}, inplace = True)
```

```python
fifa['Release Clause'] = fifa['Release Clause'].apply(clean_column_values)
fifa.rename(columns = {'Release Clause':'Release Clause(€)'}, inplace = True)
```

```python
def clean_Height(x):
    x = str(x)  # Ensure x is a string
    x = x.replace("'", '.')  # Replace . 
    x = x.replace('"', '')  # Remove "
    return x
```

```python
fifa['Height'] = fifa['Height'].apply(clean_Height)
```
